```python
from sklearn import datasets
from sklearn import model_selection
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
dataset = datasets.load_iris()
```

```python
X=dataset.data
Y=dataset.target
print(X)
print(Y)
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled=np.insert(X_scaled,0,values=1,axis=1)
```

```python
X_scaled
```

```python
X_train, X_test, Y_train, Y_test =model_selection.train_test_split(X_scaled,Y, test_size=0.25,random_state=48)
```

```python
#Part1: For one vs rest strategy, setting Y_train values to 1 for class 0 and 0 for class 1 and 2
print(Y_train)
Y_train[Y_train==0]=5
Y_train[Y_train==1]=6

Y_train1=np.zeros(len(Y_train))
for i in range(len(Y_train)):
    if(Y_train[i]==5):
        Y_train1[i]=1
    elif(Y_train[i]==6):
        Y_train1[i]=0
    else:
        Y_train1[i]=0
print(Y_train)
print(Y_train1)
```

```python
#Gradient Descent for part1
n=1000
alpha=0.001
m,k=X_train.shape
beta=np.zeros(k)
for i in range(n):
    cost_gradient=np.zeros(k)
    z=X_train.dot(beta)
    predicted=1/(1+np.exp(-z))
    difference=predicted-Y_train1
    for j in range(k):
        cost_gradient[j]=np.sum(difference.dot(X_train[:,j]))
    for j in range(k):
        beta[j]=beta[j]-(alpha/m)*cost_gradient[j]
print(beta)
```

```python
#Predicted values for part 1
Y_predict1=1/(1+np.exp(-(X_test.dot(beta))))
print(Y_predict1)
```

```python
#Part 2: For one vs rest strategy, setting Y_train values to 1 for class 1 and 0 for class 0 and 2
Y_train2=np.zeros(len(Y_train))
for i in range(len(Y_train)):
    if(Y_train[i]==5):
        Y_train2[i]=0
    elif(Y_train[i]==6):
        Y_train2[i]=1
    else:
        Y_train2[i]=0
print(Y_train)
print(Y_train2)
```

```python
#Gradient Descent for part 2
n=1000
alpha=0.01
m,k=X_train.shape
beta=np.zeros(k)
for i in range(n):
    cost_gradient=np.zeros(k)
    z=X_train.dot(beta)
    predicted=1/(1+np.exp(-z))
    difference=predicted-Y_train2
    for j in range(k):
        cost_gradient[j]=np.sum(difference.dot(X_train[:,j]))
    for j in range(k):
        beta[j]=beta[j]-(alpha/m)*cost_gradient[j]
print(beta)
```

```python
# Predicted values according to part 2
Y_predict2=1/(1+np.exp(-(X_test.dot(beta))))
print(Y_predict2)
```

```python
#Part 3: For one vs rest strategy, setting Y_train values to 1 for class 2 and 0 for class 1 and 0
Y_train3=np.zeros(len(Y_train))
for i in range(len(Y_train)):
    if(Y_train[i]==5):
        Y_train3[i]=0
    elif(Y_train[i]==6):
        Y_train3[i]=0
    else:
        Y_train3[i]=1
print(Y_train)
print(Y_train3)
```

```python
#Gradient Descent for part 3
n=1000
alpha=0.01
m,k=X_train.shape
beta=np.zeros(k)
for i in range(n):
    cost_gradient=np.zeros(k)
    z=X_train.dot(beta)
    predicted=1/(1+np.exp(-z))
    difference=predicted-Y_train3
    for j in range(k):
        cost_gradient[j]=np.sum(difference.dot(X_train[:,j]))
    for j in range(k):
        beta[j]=beta[j]-(alpha/m)*cost_gradient[j]
print(beta)
```

```python
#Predicted values according to part 3
Y_predict3=1/(1+np.exp(-(X_test.dot(beta))))
print(Y_predict3)
```

```python
Y_label=np.zeros(len(Y_predict1))
for i in range(len(Y_predict1)):
    if((Y_predict1[i]>Y_predict2[i])and(Y_predict1[i]>Y_predict3[i])):
        Y_label[i]=0
    elif ((Y_predict2[i]>Y_predict1[i])and(Y_predict2[i]>Y_predict3[i])):
        Y_label[i]=1
    elif ((Y_predict3[i]>Y_predict1[i])and(Y_predict3[i]>Y_predict2[i])):
        Y_label[i]=2
print(Y_label)
print(Y_test)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
n=1000 #number of iterations
alpha=10
m,k=X_train.shape
#print(m)
#print(k)
beta=np.zeros(k)
#print(beta)
for i in range(n):
    cost_gradient=np.zeros(k)
    z=X_train.dot(beta)
    predicted=1/(1+np.exp(-z))
    difference=predicted-Y_train
    for j in range(k):
        cost_gradient[j]=np.sum(difference.dot(X_train[:,j]))
    for j in range(k):
        beta[j]=beta[j]-(alpha/m)*cost_gradient[j]
print(beta)
```

```python
Y_predict=1/(1+np.exp(-(X_test.dot(beta))))
Y_label=np.zeros(len(Y_predict))
for i in range(len(Y_predict)):
    if(Y_predict[i]>=0.5):
        Y_label[i]=1
```

```python
TP=0
TN=0
FP=0
FN=0
Y_test=np.array(Y_test).reshape(-1,1)
for i in range(len(Y_label)):
    if(Y_test[i]==1 and Y_label[i]==1):
        TP=TP+1
    if(Y_test[i]==1 and Y_label[i]==0):
        FN=FN+1
    if(Y_test[i]==0 and Y_label[i]==1):
        FP=FP+1
    if(Y_test[i]==0 and Y_label[i]==0):
        TN=TN+1
print('True Positive: ',TP)
print('True Negative: ',TN)
print('False Positive: ',FP)
print('False Negative: ',FN)
accuracy=(TP+TN)/(TP+TN+FP+FN)
#For positive class:
precision_pos=TP/(TP+FP)
recall_pos=TP/(TP+FN)
f1_score_pos=2*precision_pos*recall_pos/(precision_pos+recall_pos)
print("Precision for Positive Class: ",precision_pos)
print("Recall for Positive Class: ",recall_pos)
print("F1 Score for Positive Class: ",f1_score_pos)
#For negative class
precision_neg=TN/(TN+FN)
recall_neg=TN/(TN+FP)
f1_score_neg=2*precision_neg*recall_neg/(precision_neg+recall_neg)
print("Precision for Negative Class: ",precision_neg)
print("Recall for Negative Class: ",recall_neg)
print("F1 Score for Negative Class: ",f1_score_neg)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
X_ridge=df_ridge.iloc[:,0:2]
Y_ridge=df_ridge.iloc[:,2]
scaler=StandardScaler()
X_scaled_ridge=scaler.fit_transform(X_ridge)
X_scaled_ridge=np.insert(X_scaled_ridge,0,values=1,axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train_ridge, X_test_ridge, Y_train_ridge, Y_test_ridge = train_test_split(X_scaled_ridge, Y_ridge, test_size=0.3, random_state=42)
```

```python
n=1000 #number of iterations
alpha=10
lamda=0.2

a,b=X_train_ridge.shape
#print(a)
#print(b)
beta_ridge=np.zeros(b)
#print(beta)
for i in range(n):
    cost_gradient_ridge=np.zeros(b)
    z_ridge=X_train_ridge.dot(beta_ridge)
    predicted_ridge=1/(1+np.exp(-z_ridge))
    difference_ridge=predicted_ridge-Y_train_ridge
    for j in range(b):
        cost_gradient_ridge[j]=np.sum(difference_ridge.dot(X_train_ridge[:,j])) 
    for j in range(b):
        beta_ridge[j]=beta_ridge[j]*(1-(lamda*alpha)/a)-(alpha/a)*cost_gradient_ridge[j]
        
print(beta_ridge)
```

```python
Y_predict_ridge=1/(1+np.exp(-(X_test_ridge.dot(beta_ridge))))
Y_label_ridge=np.zeros(len(Y_predict_ridge))
for i in range(len(Y_predict_ridge)):
    if(Y_predict_ridge[i]>=0.5):
        Y_label_ridge[i]=1
```

```python
TP=0
TN=0
FP=0
FN=0
Y_test_ridge=np.array(Y_test_ridge).reshape(-1,1)
for i in range(len(Y_label_ridge)):
    if(Y_test_ridge[i]==1 and Y_label_ridge[i]==1):
        TP=TP+1
    if(Y_test_ridge[i]==1 and Y_label_ridge[i]==0):
        FN=FN+1
    if(Y_test_ridge[i]==0 and Y_label_ridge[i]==1):
        FP=FP+1
    if(Y_test_ridge[i]==0 and Y_label_ridge[i]==0):
        TN=TN+1
print('True Positive: ',TP)
print('True Negative: ',TN)
print('False Positive: ',FP)
print('False Negative: ',FN)
accuracy_ridge=(TP+TN)/(TP+TN+FP+FN)
print("Accuracy: ",accuracy_ridge)
```
